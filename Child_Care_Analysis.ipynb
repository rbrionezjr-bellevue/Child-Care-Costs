```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the cleaned dataset
file_path = 'NDCP2022_cleaned_updated.csv'  # Update the path if the file is located elsewhere
df = pd.read_csv(file_path)
```

```python
# Ensure proper column names and data types for time analysis
df['studyyear'] = pd.to_numeric(df['studyyear'], errors='coerce')  # Convert year to numeric
```

```python
# Select relevant columns related to child care prices, regions, and time
price_columns = [
    'infants_bto5_months', 'infants_6to11_months', 'toddlers_12to17_months', 'toddlers_18to23_months', 
    'toddlers_24to29_months', 'toddlers_30to35_months', 'preschool_36to41_months', 'preschool_42to47_months', 
    'preschool_48to53_months', 'preK_54months_to_school_age', 'school_age_children']
```

```python
# Filter out rows with invalid years or all NaN price values
df = df.dropna(subset=['studyyear'])
price_data = df[['studyyear'] + price_columns]
```

```python
# Calculate average child care price per year
price_data['average_price'] = price_data[price_columns].mean(axis=1, skipna=True)
average_price_per_year = price_data.groupby('studyyear')['average_price'].mean().reset_index()
```

```python
# Plot the trend of child care prices over time using seaborn
plt.figure(figsize=(10, 6))
sns.set_style("darkgrid")
sns.lineplot(x='studyyear', y='average_price', data=average_price_per_year, marker='o', color='blue')
plt.title('Average Child Care Price Over Time', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Average Price', fontsize=14)
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Select childcare cost-related columns
childcare_cost_columns = ['preK_54months_to_school_age', 'school_age_children']

# Group by year and calculate average cost
childcare_cost_over_time = df.groupby('studyyear')[childcare_cost_columns].mean()

# Identify the two most expensive age groups
most_expensive_age_groups = childcare_cost_over_time.mean().nlargest(2).index.tolist()

# Plot area chart for the two most expensive age groups with blue hues
plt.figure(figsize=(12, 6))
plt.stackplot(childcare_cost_over_time.index, 
              childcare_cost_over_time[most_expensive_age_groups[0]], 
              childcare_cost_over_time[most_expensive_age_groups[1]], 
              labels=most_expensive_age_groups, 
              colors=['#4682B4', '#87CEFA'], alpha=0.7)

plt.title('Childcare Cost Over Time for the Two Most Expensive Age Groups')
plt.xlabel('Year')
plt.ylabel('Average Cost')
plt.legend(title="Age Groups")
plt.grid(True)
plt.show()
```

```python
# Load the dataset
file_path = 'NDCP2022_cleaned_updated_shortened (1).csv'
data = pd.read_csv(file_path)

# Select relevant columns for analysis
income_columns = ['mhi', 'mfi']

# Group by year and calculate average values
income_over_time = data.groupby('studyyear')[income_columns].mean()

# Reset index for seaborn plotting
income_over_time.reset_index(inplace=True)

# Melt data for seaborn
melted_data = income_over_time.melt(id_vars=['studyyear'], value_vars=income_columns, var_name='Metric', value_name='Amount')

# Rename columns for better understanding
melted_data['Metric'] = melted_data['Metric'].replace({
    'mhi': 'Median Household Income',
    'mfi': 'Median Family Income'
})

# Plot using seaborn
plt.figure(figsize=(12, 6))
sns.lineplot(data=melted_data, x='studyyear', y='Amount', hue='Metric', style='Metric', markers=True)

plt.title('Income Over Time')
plt.xlabel('Year')
plt.ylabel('Amount')
plt.legend(title='Metric')
plt.grid(True)
plt.show()
```

```python
# Select relevant columns for analysis
income_columns = ['mhi', 'mfi']
childcare_cost_columns = ['preK_54months_to_school_age', 'school_age_children']

# Convert necessary columns to numeric to avoid data type issues
data[income_columns + childcare_cost_columns] = data[income_columns + childcare_cost_columns].apply(pd.to_numeric, errors='coerce')

# Scale childcare costs to annual values (assuming monthly costs)
data[childcare_cost_columns] *= 12

# Sum all childcare costs across age groups to represent total childcare cost
data['total_childcare_cost'] = data[childcare_cost_columns].sum(axis=1)

# Group by year and calculate the total childcare cost and average income values
childcare_cost_over_time = data.groupby('studyyear')['total_childcare_cost'].mean()
income_over_time = data.groupby('studyyear')[income_columns].mean()

# Combine data for plotting
combined_data = income_over_time.join(childcare_cost_over_time.rename('Total Childcare Cost'))
combined_data.reset_index(inplace=True)

# Melt data for seaborn
melted_data = combined_data.melt(id_vars=['studyyear'], value_vars=income_columns + ['Total Childcare Cost'], var_name='Metric', value_name='Amount')

# Rename columns for better understanding
melted_data['Metric'] = melted_data['Metric'].replace({
    'mhi': 'Median Household Income',
    'mfi': 'Median Family Income',
    'Total Childcare Cost': 'Total Annual Childcare Cost'
})

# Plot using seaborn
plt.figure(figsize=(12, 6))
sns.lineplot(data=melted_data, x='studyyear', y='Amount', hue='Metric', style='Metric', markers=True)

plt.title('Income vs Total Childcare Cost Over Time')
plt.xlabel('Year')
plt.ylabel('Amount')
plt.legend(title='Metric')
plt.grid(True)
plt.show()
```
